##Daily Revenue

```python
%sql
SELECT DATE(OrderDate) AS order_day,
       SUM(TotalAmount) AS daily_revenue
FROM finals.amazon.amazon_clean
GROUP BY DATE(OrderDate)
ORDER BY order_day;
```

##Revenue by category

```python
%sql
SELECT Category,
       SUM(TotalAmount) AS revenue
FROM finals.amazon.amazon_clean
GROUP BY Category
ORDER BY revenue DESC;
```

##Top10 most sold products

```python
%sql
SELECT ProductName,
       SUM(TotalAmount) AS revenue
FROM finals.amazon.amazon_clean
GROUP BY ProductName
ORDER BY revenue DESC
LIMIT 10;
```

##Average revenue by country

```python
%sql
SELECT Country,
       SUM(TotalAmount) AS revenue
FROM finals.amazon.amazon_clean
GROUP BY Country
ORDER BY revenue DESC;
```
